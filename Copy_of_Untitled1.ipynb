```python
import pandas as pd
import numpy as np
np.random.seed(1)
data = {
    "Age": np.random.randint(25, 65, 100),
    "BMI": np.round(np.random.uniform(22, 36, 100), 1),
    "ALT": np.random.randint(25, 90, 100),
    "AST": np.random.randint(20, 80, 100),
    "Triglycerides": np.random.randint(120, 300, 100),
    "Cholesterol": np.random.randint(180, 320, 100),
    "Glucose": np.random.randint(85, 150, 100),
    "FattyLiver": np.random.randint(0, 2, 100)  # 0 = No, 1 = Yes
}
df = pd.DataFrame(data)
df.head(5)
```

```python
avg_second_col = df.iloc[:, 1].mean()
print(avg_second_col)
```

```python
avg_3rd_4th = df.iloc[:5, 2:4].mean()
print(avg_3rd_4th)
```

```python
avg_combined = df.iloc[:5, 2:4].values.mean()
print(avg_combined)
```

```python
row_wise_sum = df.sum(axis=1)
print(row_wise_sum)
```

```python
row_wise_avg = df.mean(axis=1)
max_avg = row_wise_avg.max()

print(max_avg)
```
